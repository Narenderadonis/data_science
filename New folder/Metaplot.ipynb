# Stack plot

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
playing=[8,5,7,8,13]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
plt.plot([],[],color='b',label='sleeping',linewidth=5)
plt.plot([],[],color='r',label='eating',linewidth=5)
plt.plot([],[],color='g',label='playing',linewidth=5)
plt.plot([],[],color='c',label='working',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['b','r','g','c'])
plt.show()
```

# Pie chart

```python
slices=[7,2,2,13]
activities=['eating','sleeping','working','playing']
col=['m','g','c','k']
plt.pie(slices,
       labels=activities,
       colors=col,
       startangle=90,
       shadow=True,
       explode=(0,0,1,0),
        autopct='%1.1f%%'
       )
plt.title("Pie chart")
plt.show()
```
